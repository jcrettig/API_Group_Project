```python
# Dependencies
import pandas as pd

#Save path to data set in a variable
data_filea = "Resources/QuickFacts Jan-30-2021a.csv"
data_fileb = "Resources/QuickFacts Jan-30-2021b.csv"
c_data_file = "Output/Coordinates.csv"
```

```python
# Use Pandas to read data from csv file "a"
dfa = pd.read_csv(data_filea)
dfa.head()
```

```python
# Use Pandas to read data from csv file 'b'
dfb = pd.read_csv(data_fileb)
dfb.head()
```

```python
# Pull in desired columns from file "a" and rename
column_dfa = dfa.loc[:, ['Fact','St. Louis city, Missouri','Austin city, Texas','Kansas City city, Missouri','Lexington city, North Carolina','Memphis city, Tennessee' , 'Lockhart city, Texas']]

#Rename Columns
column_dfa = column_dfa.rename(columns={
    'St. Louis city, Missouri' : 'St. Louis',
    'Austin city, Texas' : 'Austin',
    'Kansas City city, Missouri' : 'Kansas City',
    'Lexington city, North Carolina' : 'Lexington',
    'Memphis city, Tennessee' : 'Memphis', 
    'Lockhart city, Texas' : 'Lockhart'
})

column_dfa.head()
```

```python
# Pull in desired columns from file "b" and rename
column_dfb = dfb.loc[:, ['New York city, New York','Atlanta city, Georgia','Nashville-Davidson (balance), Tennessee', 'Dallas city, Texas', 'United States' ]]

#Rename Columns
column_dfb = column_dfb.rename(columns={
    'New York city, New York' : 'Brooklyn',
    'Atlanta city, Georgia' : 'Atlanta',
    'Nashville-Davidson (balance), Tennessee' : 'Nashville',
    'Dallas city, Texas' : 'Dallas'
    
})

column_dfb.head()
```

```python
# Merge the data from files "a" and "b" into a combined dataframe
column_df = pd.merge(column_dfa, column_dfb, left_index = True, right_index = True)
column_df.head()
```

```python
# Create a list of the desired rows to keep in the dataframe
index_list = [0, 1, 4, 5, 6, 7, 20, 23, 29, 30, 31, 35, 36, 37, 41, 44, 45, 46, 59, 60, 61]
```

```python
#Create a list of Short Fact descriptions for the rows we are keeping
SFact_df = pd.DataFrame(
    {"SFact":['Pop2019','Pop2010', "Pop%<5", 'Pop%<18','Pop%<64', 'PopFemale', 'MedValHm', 'MedRent', '%WComputer',
              '%WInternet', '%HSGrad+','%inLabForce', '%FinLabForce', 'A&FSales1K', 'RSales1K', 'MedHInc', 'PerCapInc', 
              "Poverty%", 'PopSqMile', 'LandAreaSqM', "FIPSCode"],
     "SIndex":[0, 1, 4, 5, 6, 7, 20, 23, 29, 30, 31, 35, 36, 37, 41, 44, 45, 46, 59, 60, 61]
    }
) 

SFact_df = SFact_df.set_index('SIndex')
SFact_df.head()
```

```python
#Create a new dataFrame with only the desired rows
line_df = column_df.loc[column_df.index.isin(index_list)] 
line_df = line_df.copy()
line_df.head()
```

```python
# Merge the SFact_DF and line_DF into a combined dataframe
merge_df = pd.merge(SFact_df, line_df, left_index = True, right_index = True)
merge_df.head()
```

```python
#Export to file as CSV
merge_df.to_csv("Output/CBDetailFileExpanded.csv", index=True, header = True)
```

```python
#change index column to the SFact Column and remove the Fact Column prior to transposing 
#so the SFact Colun will be the new header row
new_line_df = merge_df.set_index('SFact')
new_line_df = new_line_df[['St. Louis', 'Austin', 'Kansas City', 'Lexington','Memphis', 'Lockhart', 'Brooklyn', 'Atlanta', 'Nashville', 'Dallas']]
new_line_df.head()
```

```python
# Transpose the dataframe 
CB_df = new_line_df.transpose(copy=True)
CB_df.head()
```

```python
#Convert columns from strings to integers and floats
CB_df['Pop2019'] = [int(x.replace(",","")) for x in CB_df['Pop2019']]
CB_df['Pop2010'] = [int(x.replace(",","")) for x in CB_df['Pop2010']]
CB_df['Pop%<5'] = [float(x.replace("%","")) for x in CB_df['Pop%<5']]
CB_df['Pop%<18'] = [float(x.replace("%","")) for x in CB_df['Pop%<18']]
CB_df['Pop%<64'] = [float(x.replace("%","")) for x in CB_df['Pop%<64']]
CB_df['PopFemale'] = [float(x.replace("%","")) for x in CB_df['PopFemale']]
CB_df['MedValHm'] = [int(x.replace("$","").replace(",","")) for x in CB_df['MedValHm']]
CB_df['MedRent'] = [int(x.replace("$","").replace(",","")) for x in CB_df['MedRent']]
CB_df['%WComputer'] = [float(x.replace("%","")) for x in CB_df['%WComputer']]
CB_df['%WInternet'] = [float(x.replace("%","")) for x in CB_df['%WInternet']]
CB_df['%HSGrad+'] = [float(x.replace("%","")) for x in CB_df['%HSGrad+']]
CB_df['%inLabForce'] = [float(x.replace("%","")) for x in CB_df['%inLabForce']]
CB_df['%FinLabForce'] = [float(x.replace("%","")) for x in CB_df['%FinLabForce']]
CB_df['A&FSales1K'] = [int(x.replace("$","").replace(",","")) for x in CB_df['A&FSales1K']]
CB_df['RSales1K'] = [int(x.replace("$","").replace(",","")) for x in CB_df['RSales1K']]
CB_df['MedHInc'] = [int(x.replace("$","").replace(",","")) for x in CB_df['MedHInc']]
CB_df['PerCapInc'] = [int(x.replace("$","").replace(",","")) for x in CB_df['PerCapInc']]
CB_df['Poverty%'] = [float(x.replace("%","")) for x in CB_df['Poverty%']]
CB_df['PopSqMile'] = [float(x.replace(",","")) for x in CB_df['PopSqMile']]
CB_df['LandAreaSqM'] = [float(x.replace(",","")) for x in CB_df['LandAreaSqM']]

CB_df 
```

```python
#Verify that all columns were converted to integers and floats
CB_df.dtypes
```

```python
# Use Pandas to read data from csv file "a"
coordinates_df = pd.read_csv(c_data_file)
#Rename Columns
coordinates_df = coordinates_df.rename(columns={'City' : 'SFact'}) # rename for merge
coordinates_df = coordinates_df.set_index('SFact')
coordinates_df.head(15)
```

```python
# Merge the data from files "a" and "b" into a combined dataframe
CBDX_df = pd.merge(CB_df, coordinates_df, left_index = True, right_index = True)
CBDX_df.head()
```

```python
#Begin Final Clean Up of CBDX_df

#Convert City column from Index column
CBDX_df = CBDX_df.reset_index(drop=False)

#Rename index column to City
CBDX_df = CBDX_df.rename(columns={'index' : 'City'})
```

```python
#Continue final clean up

#Rename City Rows to agree with CityGrid Output
CBDX_df['City'] = CBDX_df['City'].replace(
    {'Kansas City': 'Kansas City, MO',
     'St. Louis' : 'St. Louis, MO',
     'Austin': "Austin, TX", 
    'Lexington': 'Lexington, NC',
     'Memphis': 'Memphis, TN',
     'Lockhart':'Lockhart, TX',
     'Brooklyn':'Brooklyn, NY',
     'Atlanta':'Atlanta, GA',
     'Nashville': 'Nashville, TN',
     'Dallas': 'Dallas, TX'
    
    })


# Elimnate a couple of unnecessary columns:
del CBDX_df['Unnamed: 0']
del CBDX_df['State']

CBDX_df.head(15)
```

```python
#Export to file as CSV
CBDX_df.to_csv("Output/CBDataFileExpanded.csv", index=True, header = True)
```
